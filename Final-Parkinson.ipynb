```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics  
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("Parkinsson disease.csv")
```

```python
df.describe()
```

```python
df.info()
```

```python
df['status'].value_counts()
```

```python
X=df.drop(columns=['name','status'],axis=1)
Y=df['status']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
scaled = StandardScaler()
scaled.fit(X_train)
X_train = scaled.transform(X_train)
X_test = scaled.transform(X_test)
```

## SVM

```python
model=svm.SVC(kernel='linear')
model.fit(X_train,Y_train)
X_test_prediction=model.predict(X_test)
accuracy_score(Y_test,X_test_prediction)
```

## LINEAR REGRESSION

```python
model = LinearRegression()
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
y_pred_class = [1 if val > 0.5 else 0 for val in y_pred]
accuracy = accuracy_score(Y_test, y_pred_class)
accuracy
```

## LOGISTIC REGRESSION

```python
log_reg=LogisticRegression().fit(X_train,Y_train)
test_pred=log_reg.predict(X_test)
train_pred=log_reg.predict(X_train)
log_test_accuracy=accuracy_score(Y_test,test_pred)
log_test_accuracy
```

## NAIVE BAYES

```python
gb=GaussianNB().fit(X_train,Y_train)
test_pred_gb=gb.predict(X_test)
accuracy_score(Y_test,test_pred_gb)
```

## KNN

```python
kn = KNeighborsClassifier(n_neighbors=3).fit(X_train, Y_train)
test_pred_kn = kn.predict(X_test)
test_accuracy = accuracy_score(Y_test, test_pred_kn)
test_accuracy
```

## DECISION TREE

```python
dt=DecisionTreeClassifier().fit(X_train,Y_train)
train_pred_dt=dt.predict(X_train)
test_pred_dt=dt.predict(X_test)
accuracy_score(Y_test,test_pred_dt)
```

## SVM KERNEL

```python
model = SVC(kernel='rbf')
model.fit(X_train, Y_train)
y_pred = model.predict(X_test)
if hasattr(model, 'probability') and not model.probability:
    model.probability = True
accuracy = accuracy_score(Y_test, y_pred)
accuracy
```

## RANDOM FOREST

```python
rfc = RandomForestClassifier(criterion='gini', max_depth=5, min_samples_leaf=5,
                       n_estimators=100, n_jobs=-1, oob_score=True,
                       random_state=42) 
rfc.fit(X_train, Y_train)
y_pred = rfc.predict(X_test)
#Evaluation Metric
print("OOB SCORE",rfc.oob_score_)
print("Accuracy", accuracy_score(Y_test, y_pred))
```

```python
#Random Forest
from sklearn.metrics import accuracy_score, roc_auc_score, roc_curve, mean_squared_error
import matplotlib.pyplot as plt
# Fit the model
rfc.fit(X_train, Y_train)
# Predict on train and test sets
y_pred_train = rfc.predict(X_train)
y_pred_test = rfc.predict(X_test)
# Calculate probabilities for ROC AUC and MSE
y_train_proba = rfc.predict_proba(X_train)[:, 1]
y_test_proba = rfc.predict_proba(X_test)[:, 1]
# Print accuracy, AUC, and MSE for the train set
train_accuracy = accuracy_score(Y_train, y_pred_train)
train_roc_auc = roc_auc_score(Y_train, y_train_proba)
train_mse = mean_squared_error(Y_train, y_pred_train)
print("The TRAIN accuracy is", train_accuracy)
print("The ROC AUC score for TRAIN data is", train_roc_auc)
print("The TRAIN Mean Squared Error is", train_mse)
print("--" * 50)
# Print accuracy, AUC, and MSE for the test set
test_accuracy = accuracy_score(Y_test, y_pred_test)
test_roc_auc = roc_auc_score(Y_test, y_test_proba)
test_mse = mean_squared_error(Y_test, y_pred_test)
print("The TEST accuracy is", test_accuracy)
print("The ROC AUC score for TEST data is", test_roc_auc)
print("The TEST Mean Squared Error is", test_mse)
```
