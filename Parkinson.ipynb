**ABSTRACT: Parkinson's disease (PD) is a progressive neurological disorder marked by motor and non-motor symptoms, such as tremors and postural instability. Early diagnosis is essential for effective treatment and management. This research reviews recent advances in machine learning (ML) techniques applied to the detection and diagnosis of PD. Various ML algorithms, including support vector machines (SVM), random forests (RF), Decision tree, etc have been utilized to analyze voice recordings. This approach has demonstrated promising results in differentiating between PD patients and healthy controls. Specifically, voice recordings are analyzed for subtle speech pattern changes. Overall, the integration of ML techniques holds great potential for enhancing early diagnosis and improving patient outcomes in Parkinson's disease.**

**DATASET: The dataset utilized in the research was obtained from the University of Oxford (UO) repository with collaboration from the National Center for Voice**

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
from sklearn.metrics import accuracy_score,  precision_score,recall_score
```

```python
X=df.drop(columns=['name','status'],axis=1)
Y=df['status']
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Train Random Forest
rfc = RandomForestClassifier(random_state=42)
rfc.fit(X, Y)

# Feature Importance
importances = rfc.feature_importances_
feature_importance_df = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
}).sort_values(by='Importance', ascending=False)

print(feature_importance_df)
```

```python
import numpy as np
import pandas as pd
import librosa

def extract_f0(audio_file):
    # Load audio file
    y, sr = librosa.load(audio_file, sr=None)
    
    # Extract fundamental frequency (F0) using librosa.pyin
    f0, voiced_flag, voiced_probs = librosa.pyin(y, fmin=librosa.note_to_hz('C2'), fmax=librosa.note_to_hz('C7'))
    
    # Remove NaNs for valid F0 values
    f0 = f0[~np.isnan(f0)]
    
    return f0

def compute_nonlinear_measures(f0):
    if len(f0) == 0:
        return None, None, None, None
    
    # Spread 1: Range of F0 values
    spread1 = np.max(f0) - np.min(f0)
    
    # Spread 2: Standard deviation of F0 values
    spread2 = np.std(f0)
    
    # PPE (Pitch Perturbation Index): Average of absolute differences between consecutive F0 values
    ppe = np.mean(np.abs(np.diff(f0))) if len(f0) > 1 else 0
    
    return spread1, spread2, ppe, f0

def compute_jitter_ddp(f0):
    if len(f0) < 3:
        return None
    
    # Calculate pitch periods
    periods = 1 / f0

    # Calculate differences of differences
    diff1 = np.diff(periods)
    diff2 = np.diff(diff1)
    
    jitter_ddp = np.mean(np.abs(diff2)) / np.mean(periods) if len(diff2) > 0 else 0
    
    return jitter_ddp

# Example usage
audio_file = 'audio/Pd8.m4a'
f0 = extract_f0(audio_file)
spread1, spread2, ppe, f0 = compute_nonlinear_measures(f0)
jitter_ddp = compute_jitter_ddp(f0)

# Calculate MDVP:Fo(Hz) and MDVP:Flo(Hz) using approximations
mdvp_fo = np.mean(f0) if len(f0) > 0 else None
mdvp_flo = np.min(f0) if len(f0) > 0 else None

# Add constant MDVP:RAP (for demonstration, this needs appropriate computation)
mdvp_rap = np.std(np.diff(f0)) / np.mean(f0) if len(f0) > 1 else None

# Prepare data for appending
row_data = {
    'PPE': round(ppe,6),
    'spread1': round(spread1,6),
    'MDVP:Fo(Hz)': round(mdvp_fo,6),
    'Jitter:DDP': round(jitter_ddp,6),
    'MDVP:RAP': round(mdvp_rap,6),
    'MDVP:Flo(Hz)': round(mdvp_flo,6),
    'spread2': round(spread2,6),
    'status': 0
}

row_data
```
